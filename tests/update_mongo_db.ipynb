# This code only updates some fields for ds005514

```python
from eegdash import EEGDash
from eegdash.data_utils import RawEEGDash

eegdashdata = EEGDash(is_public=False)
records = eegdashdata.find({'dataset': 'ds005514'}) # current call
# records = eegdashdata.find(mongodb={'dataset': 'ds005514'}, description=['age', 'gender']) # braindecode call
record = records[0]

print(eegdashdata.get_s3path(record))
```

```python
# load participants.tsv as pandas dataframe
import pandas as pd
participants_df = pd.read_csv( 'participants_ds005514.tsv', sep='\t')

# remove prefix sub- from participants_df.participant_id
participants_df['participant_id'] = participants_df['participant_id'].str.replace('sub-', '')
participants_df.head()
```

```python
# update database with age and gender

count = 0
for record in records:
    primary_key = record['data_name']

    # find age for record['subject'] from participants_df
    age = participants_df.loc[participants_df['participant_id'] == record['subject'], 'age'].values[0]
    gender = participants_df.loc[participants_df['participant_id'] == record['subject'], 'sex'].values[0]

    # update the record
    eegdashdata.update({'data_name': primary_key, #           'data_name': 'test', 
        'participantinfo': {'age': age, 'gender': gender}})
    
    # print progress
    count += 1
    print(f'Updated {count} of {len(records)} records')
```
